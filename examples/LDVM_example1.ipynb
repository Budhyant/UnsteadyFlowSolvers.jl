# Example1 - LDVM code 

### This study illustrates some kinematics with Leading Edge Vortex (LEV) formation and shedding. They appears in the following publication:
###### Ramesh, K., Gopalarathnam, A., Granlund, K., Ol, M.V., Edwards, J.R., 2014. Discrete vortex method with novel shedding criterion for unsteady airfoil flows with intermittent leading-edge vortex shedding. Journal of Fluid Mechanics 751, 500–538.

```julia
#Include the UNSflow libraries
#Pkg.clone("git://github.com/KiranUofG/UNSflow")
include("../src/UNSflow.jl")
using UNSflow
```

### Case study 1 from the paper

### Case details:
### An SD7003 airfoil undergoes a pitch-ramp-hold maneuver of amplitude 25 deg, reduced frequency 0.11 and smoothing parameter 11. 

### See earlier examples for info about the "Eldredge function" used to generate the ramp-hold kinematics. See the publication for more details. The code below shows how to implement this is UNSflow.

### Define all the kinematics.





```julia
#Define pitch angle as described above
alphadef = EldRampReturnDef(25*pi/180,0.11,11)

#Plunge magnitude is constant = 0
hdef = ConstDef(0.)

#Freestream velocity is a constant. Set as constant = 1
udef = ConstDef(1.)

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)
```

### Define the surface undergoing motion and the flowfield. The LESP_crit used below is the critical LESP parameter which is used to predict and modulate LEV shedding - it depends on the airfooil shape and Reynolds number of operation. See paper for more details.

```julia
pvt = 0.0 #leading edge

lespcrit = [0.18;] #Note that the value is entered as an array. 

# The argument "FlatPlate" provides zero camber. To use an airfoil, specify the coordinate file which should be
#in the same folder. For eg. "sd7003.dat". 
#The XFOIL convention must be used for the coordinate: TE->Upper surface->LE->Lower surface->TE
surf = TwoDSurf("sd7003_fine.dat", pvt, full_kinem,lespcrit)

curfield = TwoDFlowField()

```

### Specify simulation details and run code. 

```julia
#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 

time_total = 7. #Required total time
nsteps = Int(round(time_total/0.015))



#Now the code is run as :
#The variables mat, surf and curfield are returned
mat, surf, curfield = ldvm(surf, curfield, nsteps)
```

### Studying the results

```julia
### Let's look at the simulation results
# surf and curfield refer to the airfoil and the flowfield.
# mat contains quantites of interest -> (time,pitch,plunge,velocity,A0,Cl,Cd,Cm)

#Plot the lift coefficient
plot(mat[:,1],mat[:,6])
```

### Case study 3 from the paper 

### Case details:
### An SD7003 airfoil at a Reynolds number of 10,000 undergoesvarious sinusoidal kinematics

### 3A - alpha_mean = 4 deg, alpha_max = 22.5 deg, h/c_max = 0.05, k = 3.93, phi = 90 deg, pvt = 3/4 chord 

```julia
#Say we want to run the simulation for 5 cycles of motion.
ncyc = 5
alpha_mean = 4.*pi/180
alpha_amp = 22.5*pi/180
h_amp = 0.05
k = 3.93
phi = 90*pi/180 
T = pi/k # time period
t_tot = ncyc*T

#Define pitch and plunge as described above
alphadef = CosDef(alpha_mean, alpha_amp, k, phi)

hdef = CosDef(0., h_amp, k, 0.)

#Freestream velocity is a constant. Set as constant = 1
udef = ConstDef(1.)

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)
```

```julia
pvt = 0.75 #three-quarter chord

lespcrit = [0.21;]

surf = TwoDSurf("sd7003_fine.dat", pvt, full_kinem, lespcrit)

curfield = TwoDFlowField()

#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 

#dtstar = minimum([0.015*0.2/(k*alpha_amp) 0.015*0.2/(k*h_amp)])
dtstar = find_tstep([alphadef hdef])
nsteps = Int(round(t_tot/dtstar))

#Now the code is run as :
#The variables mat, surf and curfield are returned
mat, surf, curfield = ldvm(surf, curfield, nsteps, dtstar)
```

```julia
#Plot the lift coefficient
range = round(Int,(ncyc-1)*nsteps/ncyc)+1:nsteps
tbyT = (mat[range,1] - mat[range[1]])/T #time divided by time period 

plot(tbyT,mat[range,6])
```

### Case study 4 from the paper 

### Case details:
### A NACA0015 airfoil at a Reynolds number of 1100 undergoes sinusoidal kinematics - Based on a study performed by Kinsey and Dumas - see paper for reference

```julia
#Say we want to run the simulation for 5 cycles of motion.
ncyc = 5
alpha_mean = 0.*pi/180
alpha_amp = 76.33*pi/180
h_amp = 1.0
k = 0.377
phi = 90.0*pi/180 
T = pi/k # time period
t_tot = ncyc*T

#Define pitch and plunge as described above
alphadef = CosDef(alpha_mean, alpha_amp, k, phi)

hdef = CosDef(0., h_amp, k, 0.)

#Freestream velocity is a constant. Set as constant = 1
udef = ConstDef(1.)

#The full kinematics are collected as:
full_kinem = KinemDef(alphadef, hdef, udef)
```

### Note the DelVortDef feature used below - deletes vortices that are at a specified distance from the airfoil and may hence be neglected

```julia
pvt = 1./3. #doesn't matter here

lespcrit = [0.19;]

surf = TwoDSurf("FlatPlate", pvt, full_kinem, lespcrit)

curfield = TwoDFlowField()

#Arguments are (enable(1)/disable , minimum number of vortices required to enable deletion scheme,
#number of chordlengths from airfoil at which to delete)
del = DelVortDef(1, 500, 10)

#Specify simulation details - time step, and number of steps.
#The default nondimensional timestep is 0.015. There is usually no need to change this. 

#dtstar = minimum([0.015*0.2/(k*alpha_amp) 0.015*0.2/(k*h_amp)])
dtstar = find_tstep([alphadef hdef])
nsteps = Int(round(t_tot/dtstar))

#Now the code is run as :
#The variables mat, surf and curfield are returned
mat, surf, curfield = ldvm(surf, curfield, nsteps, dtstar, del)
```

```julia
#Plot the lift coefficient
range = round(Int,(ncyc-1)*nsteps/ncyc)+1:nsteps
tbyT = (mat[range,1] - mat[range[1]])/T #time divided by time period 

plot(tbyT,mat[range,6])
```
